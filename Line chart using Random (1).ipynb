```python
num1=np.random.randint(20,50,(20))
```

```python
num2=np.random.randint(30,60,(20))
```

```python
num2
```

```python
plt.plot(num1)
plt.plot(num2)
plt.show()
```

```python
plt.bar(num1,num2)
```

```python
plt.barh(num1,num2)
```

```python
plt.plot(num1,'b--',label="Random 1",linestyle='-',marker='+')
plt.plot(num2,'r--',label="Random 2",linestyle='-.',marker='o')
plt.title("Random data")
plt.xlabel("Random numbeer 1")
plt.ylabel("Random numbeer 2")
plt.legend(loc=4)
plt.show()
```
