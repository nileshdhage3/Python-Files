```python
#  Assignment 08/04/2021

#  Algoritham

#  1. Store product details with price (Datatype - Dictionary datatype we can use)
#  2. Discount details: total amount more than 2000 then 10% discount,
#     amount between 1000 to 1999 then 7% discount,
#     Between 500 to 999 then 5% discount,
#     below 500 50 Rs discount *(Nested if-else can be use)
#  3. Accept user input for product details 
#     (quantity and product / product ID) *(Input fuction & save the value)
#  4. perform the calculation *(List can be used to store product * qty)
#  5. Net amount to be paid by customer
#  6. Total discount amount 
#  7. Final bill (*Net amount = total bill - discount)


# User name input for record purpose
name=input("Welcome to our super market, Please enter your name and start for shopping")

# Storing item details with price
storepp={'Sugar':80,'Oil':200,'Toothpaste':100,'Salt':40,'Peanuts':120}

# User input to purchase item quantity
qtysugar=int(input("Please enter the sugar quantity to purchase"))
qtyoil=int(input("Please enter the oil quantity to purchase"))
qtytooth=int(input("Please enter the Toothpaste quantity to purchase"))
qtysalt=int(input("Please enter the Salt quantity to purchase"))
qtypeanuts=int(input("Please enter the Peanuts quantity to purchase"))

# Total amount of purchase item
pricetotal=(storepp['Sugar']*qtysugar)+(storepp['Oil']*qtyoil)+(storepp['Toothpaste']*qtytooth)+(storepp['Salt']*qtysalt)+(storepp['Peanuts']*qtypeanuts)

from datetime import date
datetm=date.today()


# Discount calculation and calculate net amount to pay
if pricetotal>=2000:
    discount=pricetotal*10/100
    
elif pricetotal>1000 and pricetotal<2000:
    discount=pricetotal*5/100
    
else:
    discount=50
nettotal=pricetotal-discount

# Purchase record print in table format
print("\n\nCustomer Name: ",name,"  ","Date:",datetm,"\n\n               --Order details-- \n\nProduct Name -- Quantity -- Price  --  Total Amount\n",
      "--------------------------------------------------",
      "\nSugar","           ",      qtysugar,"        ",storepp['Sugar'],"       ",qtysugar*storepp['Sugar'],
      "\nOil              ",qtyoil,"        ",storepp['Oil'],"      ",qtyoil*storepp['Oil'],
      "\nToothpaste       ",qtytooth,"        ",storepp['Toothpaste'],"      ",qtytooth*storepp['Toothpaste'],
      "\nSalt             ",qtysalt,"        ",storepp['Salt'],"       ",qtysalt*storepp['Salt'],
      "\nPeanuts          ",qtypeanuts,"        ",storepp['Peanuts'],"      ",qtypeanuts*storepp['Peanuts'],
      "\n------------------------------------------------",
      "\nTotal            ",qtysugar+qtyoil+qtytooth+qtysalt+qtypeanuts,"        --        ",pricetotal,
      "\nDiscount                               ",discount,
      "\n------------------------------------------------",
      "\nNet payable amount                     ",pricetotal-discount,
      "\n------------------------------------------------",
      "\nThank you for shopping, Please Visit Again")
```
