```python
#Plot the box graph for all 3 columns
station_antwerp=data['station_antwerp'].fillna(data['station_antwerp'].mean(),inplace=True)
plt.boxplot(data['station_antwerp'],vert=False)
plt.show()
```

```python
#Plot the box graph for all 3 columns
station_antwerp=data['station_paris'].fillna(data['station_paris'].mean(),inplace=True)
plt.boxplot(data['station_paris'],vert=False)
plt.show()
```

```python
#Plot the box graph for all 3 columns
station_antwerp=data['station_london'].fillna(data['station_london'].mean(),inplace=True)
plt.boxplot(data['station_london'],vert=False)
plt.show()
```

```python
plt.plot(data['station_london'],marker='.',color="blue")
plt.xlabel("station_antwerp")
plt.ylabel("Air quality")
plt.title("Air Quality - station_antwerp")
plt.show()
```

```python
plt.plot(data['station_paris'],marker='*',color="red")
plt.xlabel("station_paris")
plt.ylabel("Air quality")
plt.title("Air Quality - station_paris")
plt.show()
```

```python
plt.plot(data['station_london'],marker='>',color="green")
plt.xlabel("station_paris")
plt.ylabel("Air quality")
plt.title("Air Quality - station_london")
plt.show()
```

```python
plt.figure(figsize=(22,12))
plt.plot(data['station_london'],marker='.',color="blue",label="station_london")
plt.plot(data['station_paris'],marker='*',color="red",label="station_paris")
plt.plot(data['station_london'],marker='>',color="green",label="station_london")
plt.xlabel("station names")
plt.ylabel("Air quality")
plt.legend()
plt.title("Air Quality all 3 stations")
plt.show()
```
