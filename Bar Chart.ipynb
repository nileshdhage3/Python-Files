```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
```

```python
classes=["DS","Python","Dot Net","Java"]
class1=[10,30,50,40]
class2=[25,15,45,35]
```

```python
plt.bar(classes,class1)
```

```python
plt.barh(classes,class1)
```

```python
plt.bar(classes,class1)
plt.bar(classes,class2)
plt.show()
```

```python
#class_ind=np.arange(len(classes))
#width=0.2
style.use("ggplot")
plt.figure(figsize=(12,6))
plt.bar(classes,class1,label="class 1 stud")
plt.bar(classes,class2,label="class 2 stud")
plt.title("Class - Student")
plt.xlabel("Class")
plt.ylabel("Marks")
plt.legend()
plt.show()
```
